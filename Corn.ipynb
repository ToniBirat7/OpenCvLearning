```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt

cv.setUseOptimized(True)
```

```python
img = cv.imread('./Maize_img/m1.jpg')

rows, cols, channel = img.shape

print("Shape", img.shape)

grey_img = cv.cvtColor(img,cv.COLOR_BGR2GRAY)

# thres = cv.adaptiveThreshold(after_blur,255,cv.ADAPTIVE_THRESH_GAUSSIAN_C,cv.THRESH_BINARY_INV,5,2)
thres = cv.adaptiveThreshold(grey_img,255,cv.ADAPTIVE_THRESH_MEAN_C,cv.THRESH_BINARY_INV,7,2)
# thres = cv.threshold(after_blur,0,255,cv.THRESH_BINARY+cv.THRESH_OTSU)

after_blur = cv.GaussianBlur(thres,(5,5),1)

kernel = np.ones((3,3),np.uint8)

opening = cv.morphologyEx(thres, cv.MORPH_OPEN,kernel=kernel)

# cv.imshow("Original", img)
# cv.imshow("Grey", grey_img)
# cv.imshow("After Blur", after_blur)
# cv.imshow("Thres", thres)
# cv.imshow("After Opening",opening)
# cv.waitKey(0)
# cv.destroyAllWindows()

plt.figure(figsize=(60,80))
plt.subplot(531), plt.title("Original"), plt.imshow(cv.cvtColor(img,cv.COLOR_BGR2RGB)), plt.axis('off')
plt.subplot(532), plt.title("Grey"), plt.imshow(grey_img, cmap='gray'), plt.axis('off')
plt.subplot(533), plt.title("After Blur"), plt.imshow(after_blur, cmap='gray'), plt.axis('off')
plt.subplot(534), plt.title("Thres"), plt.imshow(thres, cmap='gray'), plt.axis('off')
plt.subplot(535), plt.title("After Opening"), plt.imshow(opening, cmap='gray'), plt.axis('off')
plt.axis('off')
```
